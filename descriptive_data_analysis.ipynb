### Aydınlatma Niteliği için Detaylı Veri Analizi

```python
data.describe()
```

```python
data["Aydınlatma"].value_counts()
```

```python
sorted_data = data["Aydınlatma"].sort_values()
sorted_data
```

#### Çeyreklikler (Quatilers) ve Çeyrekler Açıklığı (Inter-Quartile Range)  

```python
sorted_data = data["Aydınlatma"].sort_values()
q1 = sorted_data.quantile(0.25)
q3 = sorted_data.quantile(0.75)
iqr = q3 - q1
```

```python
print("Q1: ",q1,"\nQ3: ",q3,"\nIQR: ",iqr)
```

#### Aykırılıklar

```python
alt_limit = q1 - (1.5 * iqr)
ust_limit = q3 + (1.5 * iqr)
print("Alt Limit: ",alt_limit,"\nÜst Limit: ", ust_limit) 
```

### Farklı Niteliklerin veya Nitelik Kombinasyonlarının Fiyat Dağılımı
